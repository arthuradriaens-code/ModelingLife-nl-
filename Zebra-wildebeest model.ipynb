# Zebra-wildebeest model

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
Z,W = np.meshgrid(np.linspace(0,100,15),np.linspace(0,100,15))

u = W*(1.05 - 0.1*W - 0.025*Z)
v = Z*(1.1 - 0.05*Z - 0.2*W)
M = np.sqrt(u**2 + v**2)
M[ M == 0] = 1 #vermijden dat er door 0 gedeeld wordt
u /= M
v /= M
plt.quiver(Z,W,u,v,M)
plt.title('Zebra-wildebeest model')
plt.xlabel('Zebra')
plt.ylabel('Wildebeest')
plt.colorbar()
plt.show()
```
